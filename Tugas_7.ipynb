#Korelasi data numerik
Pada tugas 7 ini, kita akan mencari angka korelasi dari kolom tabel yang hanya memuat atribut kuantitatif saja. Maka pertama, kita akan lakukan korelasi satu per satu.

```python
import pandas as pd

salegame = pd.read_csv('vgsales.csv')
tulisan = 'Berdasarkan angka tersebut,  kedua kolom tabel data tersebut '
def hasil(korelasi): 
    if korelasi < -0.5:
        return tulisan + 'korelasinya berbanding terbalik'
    elif korelasi > 0.5:
        return tulisan + 'korelasinya berbanding lurus'
    else:
        return tulisan + 'tidak berkorelasi'
    
print(f"Angka korelasi antara Rank dengan Year adalah {salegame['Rank'].corr(salegame['Year'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['Year']))} \n ")

print(f"Angka korelasi antara Rank dengan NA_Sales adalah {salegame['Rank'].corr(salegame['NA_Sales'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['NA_Sales']))} \n ")

print(f"Angka korelasi antara Rank dengan EU_Sales adalah {salegame['Rank'].corr(salegame['EU_Sales'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['EU_Sales']))} \n ")

print(f"Angka korelasi antara Rank dengan JP_Sales adalah {salegame['Rank'].corr(salegame['JP_Sales'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['JP_Sales']))} \n ")

print(f"Angka korelasi antara Rank dengan Other_Sales adalah {salegame['Rank'].corr(salegame['Other_Sales'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['Other_Sales']))} \n ")

print(f"Angka korelasi antara Rank dengan Global_Sales adalah {salegame['Rank'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['Rank'].corr(salegame['Global_Sales']))} \n ")

print(f"Angka korelasi antara Year dengan NA_Sales adalah {salegame['Year'].corr(salegame['NA_Sales'])}")
print(f"{hasil(salegame['Year'].corr(salegame['NA_Sales']))} \n ")

print(f"Angka korelasi antara Year dengan EU_Sales adalah {salegame['Year'].corr(salegame['EU_Sales'])}")
print(f"{hasil(salegame['Year'].corr(salegame['EU_Sales']))} \n ")

print(f"Angka korelasi antara Year dengan JP_Sales adalah {salegame['Year'].corr(salegame['JP_Sales'])}")
print(f"{hasil(salegame['Year'].corr(salegame['JP_Sales']))} \n ")

print(f"Angka korelasi antara Year dengan Other_Sales adalah {salegame['Year'].corr(salegame['Other_Sales'])}")
print(f"{hasil(salegame['Year'].corr(salegame['Other_Sales']))} \n ")

print(f"Angka korelasi antara Year dengan Global_Sales adalah {salegame['Year'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['Year'].corr(salegame['Global_Sales']))} \n ")

print(f"Angka korelasi antara NA_Sales dengan EU_Sales adalah {salegame['NA_Sales'].corr(salegame['EU_Sales'])}")
print(f"{hasil(salegame['NA_Sales'].corr(salegame['EU_Sales']))} \n ")

print(f"Angka korelasi antara NA_Sales dengan JP_Sales adalah {salegame['NA_Sales'].corr(salegame['JP_Sales'])}")
print(f"{hasil(salegame['NA_Sales'].corr(salegame['JP_Sales']))} \n ")

print(f"Angka korelasi antara NA_Sales dengan Other_Sales adalah {salegame['NA_Sales'].corr(salegame['Other_Sales'])}")
print(f"{hasil(salegame['NA_Sales'].corr(salegame['Other_Sales']))} \n ")

print(f"Angka korelasi antara NA_Sales dengan Global_Sales adalah {salegame['NA_Sales'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['NA_Sales'].corr(salegame['Global_Sales']))} \n ")

print(f"Angka korelasi antara EU_Sales dengan JP_Sales adalah {salegame['EU_Sales'].corr(salegame['JP_Sales'])}") 
print(f"{hasil(salegame['Rank'].corr(salegame['NA_Sales']))} \n")

print(f"Angka korelasi antara EU_Sales dengan Other_Sales adalah {salegame['EU_Sales'].corr(salegame['Other_Sales'])}")
print(f"{hasil(salegame['EU_Sales'].corr(salegame['JP_Sales']))} \n")

print(f"Angka korelasi antara EU_Sales dengan Global_Sales adalah {salegame['EU_Sales'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['EU_Sales'].corr(salegame['Global_Sales']))} \n")

print(f"Angka korelasi antara JP_Sales dengan Other_Sales adalah {salegame['JP_Sales'].corr(salegame['Other_Sales'])}")
print(f"{hasil(salegame['JP_Sales'].corr(salegame['Other_Sales']))} \n")

print(f"Angka korelasi antara JP_Sales dengan Global_Sales adalah {salegame['JP_Sales'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['JP_Sales'].corr(salegame['Global_Sales']))} \n")

print(f"Angka korelasi antara Other_Sales dengan Global_Sales adalah {salegame['Other_Sales'].corr(salegame['Global_Sales'])}")
print(f"{hasil(salegame['Other_Sales'].corr(salegame['Global_Sales']))} \n")
```

#Membuat tabel berisi korelasi
Setelah mengetahui korelasi antar kolom-kolom tabel tersebut, berikut akan ditunjukkan tabel berisi angka korelasi dari masing-masing tabel beratribut kuantitatif

```python
salegame.corr(numeric_only=True)
```

#Visualisasi data
Dengan didapatkannya data di atas, dapat kita buat grafik visual angka korelasi data tersebut

```python
import seaborn as sns

sns.heatmap(salegame.corr(numeric_only=True), annot=True, cmap='twilight_shifted')
```

Dengan begitu, kita telah mendapatkan angka korelasi antar data-datanya secara tabel dan juga secara grafik berwarna.
