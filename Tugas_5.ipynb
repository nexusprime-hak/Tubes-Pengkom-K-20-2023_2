```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_1 = pd.read_csv('vgsales.csv')
df_1 = df_1.drop(columns="Rank")
```

<h1>Sampel Data</h1>
<p>Ada banyak metode sampling yang dapat digunakan dalam mengolah suatu data. Dalam tugas kali ini, kami akan menggunakan beberapa metode berikut:</p>
<ol>
<li>Systematic Sampling</li>
    <p>Systematic sampling mengelompokkan datapoints dengan suatu sistematika tertentu. Misalnya 10 data pertama, data terbesar, dan data terkecil.</p>
<li>Stratified Sampling</li>
    <p>Stratified Sampling mengelompokkan setiap datapoints ke dalam tiap-tiap strata. Strata yang dipakai ditentukan oleh setiap dataframe</p>
</ol>

<h3>Systematic Sampling: Dataframe</h3>
<p>Pada sampling ini, akan diberikan data 10 game pertama dengan penjualan global paling banyak</p>

```python
max_10 = df_1.sort_values(by='Global_Sales', ascending=False)
max_10.head(10)
```

<p>Pada sampling ini akan diberikan 10 data pertama game dengan penjualan global paling sedikit</p>

```python
min_10 = df_1.sort_values(by='Global_Sales', ascending=True)
min_10.head(10)
```

Selain itu kita juga dapat mengambil sampel secara random.

```python
sample = df_1.sample(n=30)
sample
```

<h3>Stratified Sampling: Dataframe 1</h3>
<h4>Dataframe 1 akan dikelompokkan ke dalam beberapa strata</h4>

<h5>Strata 1: Platform game</h5>

<p>Dari kode di bawah dapat diperoleh distribusi frekuensi platform atau konsol game yang dimiliki oleh para konsumen game. Diketahui bahwa platform game terbanyak adalah Nintendo DS dan yang paling sedikit adalah SEGA Game Gear.</p>

```python
platform = df_1.sort_values(by="Platform")
platform_strat = platform.groupby(by="Platform")
platform_strat.count().sort_values(by="Name", ascending=False).drop(columns=["Year","Genre","Publisher","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]).rename(columns={"Name":"Console_Count"})
```

Kita dapat melakukan systematic sampling dengan mengambil 10 sampel data dari setiap pengelompokkan genre.

```python
platform_strat.sample(n=10, replace=True)
```

<h5>Strata 2: Genre game berdasarkan jumlah penjualannya</h5>

Pertama kita kelompokkan data berdasarkan genre game.

```python
genre = df_1[["Genre","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
genre_strat = genre.groupby(by="Genre")
```

Distribusi frekuensi untuk setiap genre berdasarkan dataframe:

```python
genre_strat.count().filter(items=["Global_Sales"]).rename(columns={"Global_Sales":"Games_Count"})
```

Lalu, data yang sudah terkelompokkan dapat kita sampling dengan menggunakan systematic sampling. Misal kita ambil 10 data untuk setiap genre sebagai sampling berdasarkan grouping yang telah kita buat.

```python
genre_sample = genre_strat.sample(n=10)
genre_sample
```

<h5>Strata 3: Publisher Game</h5> <br>
Data dikelompokkan berdasarkan publisher dari game tersebut. Lalu jumlah salinan setiap game yang dibuat oleh studio game tersebut kita jumlahkan dengan method sum() untuk meninjau berapa banyak salinan game yang telah dijual oleh setiap studio tersebut.

```python
pub = df_1[["Publisher","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
pub = pub.groupby(by="Publisher")
```

Setelah itu kita dapat melihat beberapa data statistik, seperti:

Distribusi frekuensi studio game:

```python
# distribusi frekuensi
pub.count().filter(items=["Global_Sales"]).rename(columns={"Global_Sales":"Games_Count"})
```

<hr>

10 studio dengan rilisan game paling banyak berdasarkan dataframe:

```python
pub_10_top = pub.count().filter(items=["Global_Sales"]).sort_values(by="Global_Sales", ascending=False).head(10)
pub_10_top.rename(columns={"Global_Sales":"Games_Count"})
```

<hr>

10 studio dengan rilisan game paling sedikit berdasarkan dataframe

```python
pub_10_bottom = pub.count().filter(items=["Global_Sales"]).sort_values(by="Global_Sales", ascending=False).tail(10)
pub_10_bottom.rename(columns={"Global_Sales":"Games_Count"})
```

<hr>

10 studio game dengan penjualan game (secara akumulatif) paling sedikit

```python
pub.sum().sort_values(by="Global_Sales",ascending=False).tail(10)
```

<hr><br>
10 studio game  dengan penjualan game terbanyak

<h2>Statistik Data</h2>

Mari kita tinjau statistik untuk setiap strata yang sudah kita buat di atas.<br><br>
1. Strata 1: Platform Game<br>
<hr>

1.1 Rata-rata dan standar deviasi

```python
platform_mean = platform.drop(columns=['Year','Name','Genre','Publisher'])
platform_mean.groupby(by="Platform").mean().sort_values(by="Global_Sales", ascending=False)
```

Data di atas menunjukkan rata-rata jumlah game yang terjual di setiap region yang dikelompokkan berdasarkan platformnya.

```python
platform_std = platform.drop(columns=['Year','Name','Genre','Publisher'])
platform_std.groupby(by="Platform").std().dropna(axis=0, how='all')
```

Data di atas menunjukkan standar deviasi dari jumlah game yang terjual di setiap region yang dikelompokkan berdasarkan platformnya.

1.2 Ekstremum

```python
platform_max = platform.drop(columns=['Year','Name','Genre','Publisher'])
platform_max.max(axis=0, skipna=True)
```

Dari data di atas, console dengan jumlah game yang terjual paling banyak adalah XBox One.

```python
platform_min = platform.drop(columns=['Year','Name','Genre','Publisher'])
platform_min.min(axis=0, skipna=True)
```

Dari data di atas, dapat diketahui bahwa console dengan jumlah game terjual paling sedikit adalah Atari 2600.

1.3 Persentil

```python
platform_quant = platform.drop(columns=['Year','Name','Genre','Publisher']).sort_values(by="Global_Sales")
platform_quant=platform_quant.groupby(by='Platform').quantile([.1,.25,.5,.75,.9], interpolation='nearest')
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(platform_quant)
```

```python
platform_quant.groupby(by='Platform').quantile(.25, interpolation='nearest')
```

```python
platform_quant.groupby(by='Platform').quantile(.5, interpolation='nearest')
```

```python
platform_quant.groupby(by='Platform').quantile(.75, interpolation='nearest')
```

```python
platform_quant.groupby(by='Platform').quantile(.9, interpolation='higher')
```
