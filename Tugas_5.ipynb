```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_1 = pd.read_csv('vgsales.csv')
df_1 = df_1.drop(columns="Rank")
```

<h1>Sampel Data</h1>
<p>Ada banyak metode sampling yang dapat digunakan dalam mengolah suatu data. Dalam tugas kali ini, kami akan menggunakan beberapa metode berikut:</p>
<ol>
<li>Systematic Sampling</li>
    <p>Systematic sampling mengelompokkan datapoints dengan suatu sistematika tertentu. Misalnya 10 data pertama, data terbesar, dan data terkecil.</p>
<li>Stratified Sampling</li>
    <p>Stratified Sampling mengelompokkan setiap datapoints ke dalam tiap-tiap strata. Strata yang dipakai ditentukan oleh setiap dataframe</p>
</ol>

<h3>Systematic Sampling: Dataframe 1</h3>
<p>Pada sampling ini, akan diberikan data 10 game pertama dengan penjualan global paling banyak</p>

```python
max_10 = df_1.sort_values(by='Global_Sales', ascending=False)
max_10.head(10)
```

<p>Pada sampling ini akan diberikan 10 data pertama game dengan penjualan global paling sedikit</p>

```python
min_10 = df_1.sort_values(by='Global_Sales', ascending=True)
min_10.head(10)
```

<h3>Stratified Sampling: Dataframe 1</h3>
<h4>Dataframe 1 akan dikelompokkan ke dalam beberapa kelompok</h4>

<h5>Strata 1: Distribusi frekuensi platform game</h5>

<p>Dari kode di bawah dapat diperoleh distribusi frekuensi platform atau konsol game yang dimiliki oleh para konsumen game. Diketahui bahwa platform game terbanyak adalah Nintendo DS dan yang paling sedikit adalah SEGA Game Gear.</p>

```python
platform = df_1.sort_values(by="Platform")
platform = platform.value_counts(subset="Platform")
platform
```

<h5>Strata 2: Genre game berdasarkan jumlah penjualannya</h5>

Pertama kita kelompokkan data berdasarkan genre game.

```python
genre = df_1[["Genre","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
genre = genre.groupby(by="Genre")
```

Jumlah rilisan game untuk setiap genre berdasarkan dataframe:

```python
genre.count().filter(items=["Global_Sales"])
```

Jumlah salinan game yang terjual untuk setiap genrenya pada setiap region:

```python
genre.sum()
```

<h5>Strata 3: Publisher Game</h5> <br>
Data dikelompokkan berdasarkan publisher dari game tersebut. Lalu jumlah salinan setiap game yang dibuat oleh studio game tersebut kita jumlahkan dengan method sum() untuk meninjau berapa banyak salinan game yang telah dijual oleh setiap studio tersebut.

```python
pub = df_1[["Publisher","NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"]]
pub = pub.groupby(by="Publisher")
pub = pub.sum()
```

Setelah itu kita dapat melihat beberapa data statistik, seperti:

```python
# 10 studio game dengan penjualan game paling sedikit
pub.sort_values(by="Global_Sales",ascending=False).tail(10)
```

Dari hasil di atas dapat disimpulkan bahwa 10TACLE Studios menjadi studio game dengan nilai penjualan terkecil.

```python
# 10 studio game  dengan penjualan game terbanyak
pub.sort_values(by="Global_Sales",ascending=False).head(10)
```
