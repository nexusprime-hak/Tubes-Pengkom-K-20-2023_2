```python
import pandas as pd
import numpy as np

df = pd.read_csv('vgsales.csv')
```

--------------------------------

<h3>Perbandingan Kategori: </h3> <br>
Perbandingan jumlah Global Sales untuk sepuluh Publisher teratas

```python
# Perbandingan jumlah Global Sales untuk sepuluh Publisher teratas
gs10 = df[['Publisher','Global_Sales']].groupby(['Publisher']).sum().sort_values(by='Global_Sales',ascending=False).head(10)

f = gs10.plot(
                kind='pie',
                y='Global_Sales',
                figsize=(12,10), 
                autopct='%1.0f%%', 
                explode=tuple([0.01 for i in range(len(gs10.values))]),
                title='Top 10 Company Global Sales'
                ).legend(loc='upper center')
f.figure.set_facecolor('white')    
```

```python
f = gs10.plot(kind='barh',y='Global_Sales', title='Top 10 Company Global Sales',ylabel='')
f.set_xlabel('Global Sales (in Million Copies)')
```

<h3> INSIGHT </h3>
insight yang didapat dari kedua grafik diatas adalah bahwa 3 perusahaan game terbaik di dunia adalah Nintendo, diikuti dengan Elektronic Arts, dan Ativision Blizzard

--------------------------------

<h3>Perubahan Terhadap Waktu</h3>
Nintendo Sales from 1983-2016

```python
NTDsales = df.loc[df['Publisher'] == 'Nintendo'].sort_values(by='Year').dropna().groupby(by='Year').sum().drop(columns='Rank')
NTDsales.plot(kind='line',title='Nintendo Sales 1983-2016',ylabel='Sales (in Million Copies)')
```

```python
NTDsales.plot(kind='area',stacked=True,title='Nintendo Sales 1983-2016', ylabel='Sales (in Million Copies)')
```

<h3>INSIGHT</h3>
Dari grafik terlihat bahwa perusahaan game Nintende mengalami penjualan game terbaik pada tahun 2005-2010 sementara penjualan terendah terjadi pada tahun 1987

--------------------------------

<h3>Hierarki dan hubungan keseluruhan-bagian</h3>


Pembagian penjualan game dari perusahaan Nintendo

```python
#Total Sales
totalSales = df.groupby('Publisher').sum().drop(columns=['Rank','Year']).sort_values(by='Global_Sales',ascending=False)

f = totalSales.sum().drop('Global_Sales').plot(
                kind='pie',
                y='Nintendo',
                figsize=(12,10), 
                autopct='%1.0f%%', 
                explode=tuple([0.01 for i in range(len(totalSales.sum().drop('Global_Sales').values))]),
                title="Nintendo Games Sales",
                fontsize=20
                ).legend(loc='upper right')
f.figure.set_facecolor('white')
```

```python
totalSales.drop(columns='Global_Sales').head(5).plot(kind='bar',stacked=True,title='Top 5 Game Company Sales')
```

<h3>INSIGHT</h3>
Dari grafik diatas dapat disimpulkan bahwa penjualan terbesar berada di Amerika Utara dan Eropa, lalu diikuti jepang dan area lain
